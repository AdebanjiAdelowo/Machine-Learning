```python
import pandas as pd

dataset = pd.read_csv('births.csv')

dataset.head()
```

```python
dataset.day = dataset.day.fillna(0) 

dataset.day = dataset.day.astype(int)
```

```python
dataset.head()
```

```python
dataset['decade'] = 10 * (dataset['year'] // 10)

birth_pivot_table = dataset.pivot_table('births', index='decade', columns='gender', aggfunc='sum')

birth_pivot_table
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set() 
birth_decade = dataset.pivot_table('births', index='decade', columns='gender', aggfunc='sum') 
birth_decade.plot() 
plt.ylabel("Total births per year") 
plt.show()
```
